```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
im0=np.loadtxt('E:/ML_IP/ML_projects/class/lap/im0.txt')
im1=np.loadtxt('E:/ML_IP/ML_projects/class/lap/im1.txt')
im2=np.loadtxt('E:/ML_IP/ML_projects/class/lap/im2.txt')
im3=np.loadtxt('E:/ML_IP/ML_projects/class/lap/im3.txt')

lap_imgs=[im0,im1,im2,im3]
sigmas=[1,2,3,4]
```

```python
# loaded images -> computed using "Blob_Detection.ipynb"
f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, sharey=True,figsize=(20,10))
ax1.imshow(im0,cmap='gray')
ax1.set_title('blob_detection :sigma=1')
ax2.imshow(im1,cmap='gray')
ax2.set_title('blob_detection :sigma=2')
ax3.imshow(im2,cmap='gray')
ax3.set_title('blob_detection :sigma=3')
ax4.imshow(im3,cmap='gray')
ax4.set_title('blob_detection :sigma=4')
```

```python
col=cv2.imread('E:/ML_IP/image_processing/images/wonder.jpg')
#lap_imgs=[new,new_2]
for i in range(len(sigmas)):
    thr=np.max(lap_imgs[i]/1.5)
    #thr=th_new
    img=cv2.cvtColor(col, cv2.COLOR_BGR2GRAY)
    thres=(lap_imgs[i]>thr)
    sols=np.argwhere(thres==1)
    #radius of blobs are taken as 5*actual_radius
    radius=5*((2**0.5)*sigmas[i])
    r,g,b=list(map(int,(np.random.randint(0,255,(1,3))[0]).astype('uint8')))
    for point in sols:
        cv2.circle(col,(point[1],point[0]),int(radius),(r,g,b),2)
col=cv2.cvtColor(col, cv2.COLOR_RGB2BGR)
plt.figure(figsize=(10,10))
plt.imshow(col)
plt.title("Feature Detection- ScaleSpaceExtrema")
```
