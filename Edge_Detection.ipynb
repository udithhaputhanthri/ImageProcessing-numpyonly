```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def conv2D(img,kernal):
    img_conv=np.zeros(np.array(img.shape))
    n=kernal.shape[0]
    for i in range(n//2,img.shape[0]-n//2):
        for j in range(n//2,img.shape[1]-n//2):
            pixel=0
            #print('image_point :',(i,j))
            for k in range(-(n//2),n//2+1):
                for m in range(-(n//2),n//2+1):
                    #print('neighbourhood :',(i+k,j+m))
                    #print(n//2+k,n//2+m)
                    pixel+=img[i+k][j+m]*kernal[n//2+k][n//2+m]
            img_conv[i][j]=pixel          
    return img_conv

def gaussian(sigma,n=0): #dont give n as an input (n is for generate DoG)
    if n==0:
        n=int((3*sigma)%2==0)+3*sigma
    kernal=np.zeros((n,n))
    for i in range(-(n//2),n//2+1):
        for j in range(-(n//2),n//2+1):
            val=np.exp(-(i**2+j**2)/(2*sigma**2))
            kernal[i+n//2][j+n//2]=val
    return (kernal)/np.sum(kernal)

def Divergence(sigma,n=0): #dont give n as an input (n is for generate DoG)
    if n==0:
        n=int((4*sigma)%2==0)+4*sigma
    kernal=np.zeros((n,n))
    for i in range(-(n//2),n//2+1):
        for j in range(-(n//2),n//2+1):
            val=i*np.exp(-(i**2+j**2)/(2*sigma**2))+j*np.exp(-(i**2+j**2)/(2*sigma**2))
            kernal[i+n//2][j+n//2]=val
    return (kernal)
```

```python
kernal=Divergence(30)
plt.plot(kernal[kernal.shape[0]//2+1])
plt.title('Divergence of Gaussian')
```

```python
col=cv2.imread('E:/ML_IP/image_processing/images/wonder.jpg')
img=cv2.cvtColor(col, cv2.COLOR_BGR2GRAY)
```

```python
# Edge Detection using Divergence of Gaussian
kernal=Divergence(1)
edges=conv2D(img,kernal)
```

```python

ims=[img,edges]
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,10))
ax1.imshow(img,cmap='gray')
ax1.set_title('Image')
ax2.imshow(edges,cmap='gray')
ax2.set_title('Edges- Div(G)')
```

```python
#vertical and horizontal edge detection using sobel filter
vertical_kernal=np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
horizontal_kernal=np.array([[1,2,1],[0,0,0],[-1,-2,-1]])

ver_edges=conv2D(img,vertical_kernal)
hor_edges=conv2D(img,horizontal_kernal)
```

```python
ims=[ver_edges,hor_edges]
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,10))
ax1.imshow(hor_edges,cmap='gray')
ax1.set_title('Horizontal Edges')
ax2.imshow(ver_edges,cmap='gray')
ax2.set_title('Vertical Edges')
```
