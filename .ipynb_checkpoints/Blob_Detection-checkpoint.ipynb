```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def conv2D(img,kernal):
    img_conv=np.zeros(np.array(img.shape))
    n=kernal.shape[0]
    for i in range(n//2,img.shape[0]-n//2):
        for j in range(n//2,img.shape[1]-n//2):
            pixel=0
            #print('image_point :',(i,j))
            for k in range(-(n//2),n//2+1):
                for m in range(-(n//2),n//2+1):
                    #print('neighbourhood :',(i+k,j+m))
                    #print(n//2+k,n//2+m)
                    pixel+=img[i+k][j+m]*kernal[n//2+k][n//2+m]
            img_conv[i][j]=pixel          
    return img_conv 

def laplacian(sigma): #use n as odd number
    n=2*(int(int(3*sigma+1)%2==0)+int(3*sigma+1))+1
    kernal=np.zeros((n,n))
    for i in range(-(n//2),n//2+1):
        for j in range(-(n//2),n//2+1):
            val=(i**2+j**2-2*sigma**2)*np.exp(-(i**2+j**2)/(2*sigma**2))
            kernal[i+n//2][j+n//2]=val
    return (kernal)/np.sum(kernal) 
```

```python
kernal=laplacian(30)
plt.plot(kernal[kernal.shape[0]//2+1])
plt.title('Laplacian of Gaussian')
```

```python
col=cv2.imread('E:/ML_IP/image_processing/images/wonder.jpg')
img=cv2.cvtColor(col, cv2.COLOR_BGR2GRAY)
```

```python
# Edge Detection using Divergence of Gaussian
kernal=laplacian(1)
blobs=conv2D(img,kernal)
```

```python
extremas=blobs>np.max(blobs)/2
```

```python
ims=[img,blobs,extremas]
f, (ax1, ax2,ax3) = plt.subplots(1, 3, sharey=True,figsize=(15,10))
ax1.imshow(img,cmap='gray')
ax1.set_title('Image')
ax2.imshow(blobs,cmap='gray')
ax2.set_title('Blob Detection')
ax3.imshow(extremas,cmap='gray')
ax3.set_title('Extreme Points')
```
